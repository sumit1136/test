```
from gensim.models import Word2Vec
story=open("cs.txt","r")
story=story.read()
#print(story)
```

```
import re
data=re.sub(r'[\W]',' ',story)
data=re.sub(r',',' ',data)
data=data.lower()
data=re.sub(r'[\d]',' ',data)
data=re.sub(r'\s+',' ',data)
data=re.sub('men','man',data)
data=re.sub('kings','king',data)
print(data)
```

```
from nltk.tokenize import sent_tokenize,word_tokenize
import nltk
nltk.download('punkt')
sentences=nltk.sent_tokenize(data)
sentences=[word_tokenize(sentence) for sentence in sentences]
print(sentences)
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=list(stopwords.words("english"))
s=['may','let','say','hath','yet','us','ought','take','shall','one','would','could','neither']
stop_words=stop_words+s
stop_words=set(stop_words)
print(stop_words)
```

```
from nltk.stem import WordNetLemmatizer,PorterStemmer
from nltk.corpus import wordnet
nltk.download('wordnet')
for i in range(len(sentences)):
  sentences[i]=[WordNetLemmatizer().lemmatize(word,pos=wordnet.VERB) for word in sentences[i] if word not in stop_words]
  #sentences[i]=[PorterStemmer().stem(word) for word in sentences[i]]
  sentences[i]=[word for word in sentences[i]]
print(sentences)
```

```
from gensim.models import Word2Vec
model1=Word2Vec(sentences,min_count=1)
model1.wv.vocab
similar=model1.wv.most_similar('country')
print(similar)
similarity=model1.wv.similarity('country','britain')
print(similarity*100,"%")
```
